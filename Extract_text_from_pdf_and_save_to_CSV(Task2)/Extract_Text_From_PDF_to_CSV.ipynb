```python
import pdfplumber
import pandas as pd
from collections import namedtuple
```

```python
text_data = namedtuple('text_data', 'Introduction' 'Contact' 'Summary' 'Top_Skills' 'Certifications' 'Experience' 'Education')
```

```python
file_path = "E:\AI_Champ_Challenge\Linkdin_Connection_pdfs(Task1)"
file_name = "Profile"
```

```python
text_line = []
for num_files in range(1, 51):
    with pdfplumber.open(file_path + '\\' + file_name + str(num_files) + '.pdf') as pdf:
        pages = pdf.pages
        for page in pages:
            text = page.extract_text()
            for line in text.split('\n'):
                #print(line)
                text_line.append(line)
```

```python
df = pd.DataFrame(text_line)
df.head()
```
